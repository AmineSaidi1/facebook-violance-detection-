```python
#choisir une page specifique. On peut la faire d'une facon plus dynamique dans la phase de deploiment
#de preference de mettre un compteur pour cette cellule ou bien, bien choisir la page a extracter les donnée parce que ca peut 
#prendre beaucoup de temp
driver.get('https://www.facebook.com/hashtag/harassment')
while True:
    soup=BeautifulSoup(driver.page_source,"html.parser")
    all_posts=soup.find('div').getText()
```

```python
#declarer un DF semi-structuré pour placer les dennée extractée
import pandas as pd
df=pd.DataFrame(columns={'data','toxicity','insult',})
```

```python
#insert data in 'data' column while using split
for p in all_posts.split('.'):
    df.loc[len(df)]=[p,0,0]
```

```python

df
```

```python
df=df.dropna()
```

```python
#changer les types des attribue a string pour pouvoir faire les affectations
df = df.applymap(str)
```

```python
#extraire le degré de toicité avec detoxify
for i in range(len(df)):
    (df['toxicity'].iloc[[i]])=(Detoxify('original').predict(df['data'].tolist()[i])['toxicity'])
```

```python
#meme chose pour les insultes
for i in range(len(df)):
    (df['insult'].iloc[[i]])=(Detoxify('original').predict(df['data'].tolist()[i])['insult'])
```
